```python
# Import Dependencies
import pandas as pd
```

```python
CO2_csv = "Resources/owid-co2-data.csv"
co2_df = pd.read_csv(CO2_csv)

co2_df.tail(10)
```

```python
# reduced file to country, year and co2
reduced_df = co2_df.loc[:,["country", "year", "co2"]]

reduced_df.head(10)
```

```python
# filter year
reduced_df = reduced_df.loc[(reduced_df["year"] >= 1991) & (reduced_df["year"] <= 2022)]

reduced_df.tail(10)
```

```python
reduced_df.count()
```

```python
reduced_co2 = reduced_df.dropna(how='any')
```

```python
reduced_co2.count()
```

```python
reduced_co2.dtypes
```

```python
reduced_co2 = reduced_co2.rename(columns={'country': 'Country', 'year': 'Reading_date', 'co2': 'CO2_emission'})
reduced_co2
```

```python
reduced_co2.to_csv('Resources/reduced_co2.csv', index=False)
```
